```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
import time
```

```python
filename='//ZMN-HIVE/User-Data/Maria/check_registration/control/fish20_6dpf_medium_aligned.h5'
with h5py.File(filename, "r") as f:
            # List all groups
    print("Keys: %s" % f.keys())
    start=time.time()
    reg_fish_mahsa=f['data'][:,10,:,:]
    end=time.time()
    print('Time to load file: ',end-start)
    print(reg_fish_mahsa.shape)
```

```python
mean_mahsa=np.mean(reg_fish_mahsa,axis=0)
plt.imshow(mean_mahsa)
```

```python
def compute_im_corr(mean_im,time_s):
    corr_lst=[]
    for time_p in range(time_s.shape[0]):
        corr_lst.append(np.corrcoef(time_s[time_p,:,:].flatten(),mean_im.flatten())[0,1])
    return corr_lst
```

```python
corr_lst=compute_im_corr(mean_mahsa,reg_fish_mahsa)
```

```python
plt.plot(corr_lst)
```

```python
del reg_fish_mahsa
```
